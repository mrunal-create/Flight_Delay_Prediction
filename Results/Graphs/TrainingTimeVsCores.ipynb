```python
import matplotlib.pyplot as plt

# Data
cores = [1, 2, 3, 4]
training_time = [262, 90, 67, 56]

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.plot(cores, training_time, marker='o', linestyle='-', color='b', label='Training Time')

# Adding title and labels
plt.title('Training Time vs Number of Cores (XGBoost)\nPartition Size = 100, Threads = 2', fontsize=14)
plt.xlabel('Number of Cores', fontsize=12)
plt.ylabel('Training Time (Seconds)', fontsize=12)

# Adding grid
plt.grid(True)

# Display the plot
plt.legend()
plt.show()
```
