```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from experiment_results.csv
data = pd.read_csv('experiment_results.csv')

# 1. Plot Cleaning Time and Training Time vs Partition Size
plt.figure(figsize=(10, 6))
plt.plot(data['partition_size'], data['cleaning_time'], label='Cleaning Time', marker='o')
plt.plot(data['partition_size'], data['training_time'], label='Training Time', marker='o')
plt.xlabel('Partition Size')
plt.ylabel('Time (seconds)')
plt.title('Cleaning Time and Training Time vs Partition Size')
plt.legend()
plt.grid(True)
plt.show()

# 2. Plot Training Time vs Executor Cores and Executor Instances
plt.figure(figsize=(10, 6))
plt.plot(data['executor_cores'], data['training_time'], label='Training Time vs Executor Cores', marker='o')
plt.plot(data['executor_instances'], data['training_time'], label='Training Time vs Executor Instances', marker='o')
plt.xlabel('Executor Cores and Instances')
plt.ylabel('Training Time (seconds)')
plt.title('Training Time vs Executor Cores and Executor Instances')
plt.legend()
plt.grid(True)
plt.show()

# 3. Plot Cleaning Time and Training Time vs Task CPUs
plt.figure(figsize=(10, 6))
plt.plot(data['task_cpus'], data['cleaning_time'], label='Cleaning Time', marker='o')
plt.plot(data['task_cpus'], data['training_time'], label='Training Time', marker='o')
plt.xlabel('Task CPUs')
plt.ylabel('Time (seconds)')
plt.title('Cleaning Time and Training Time vs Task CPUs')
plt.legend()
plt.grid(True)
plt.show()
```
